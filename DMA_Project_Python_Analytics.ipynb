```python
mydb = sql.connect(host="localhost", user="root", password="xxxx", database="project_data")
mycursor=mydb.cursor()
mycursor.execute('select prod_type as "Product Category", round(sum(prod_quantity) * sum(prod_price),2) as "Categ Value" from Project_Data.Product group by prod_type')
result=mycursor.fetchall

Product_Category =[]
Categ_Value=[]
for i in mycursor:
    Product_Category.append(i[0])
    Categ_Value.append(i[1])

print(Product_Category)
print(Categ_Value)

def addlabels(Product_Category, Categ_Value):
    for i in range(len(Product_Category)):
        plt.text(i, Categ_Value[i], Categ_Value[i], ha = 'center')

plt.figure(figsize=(12,9))
plt.bar(Product_Category, Categ_Value)
addlabels(Product_Category, Categ_Value)
plt.xlabel("Product Category")
plt.ylabel("Value")
plt.title("Inventory Value")
plt.show()
```
